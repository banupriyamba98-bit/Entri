```python
## Numpy Array Operations:
```

```python
### 1. Create a 1D NumPy Array:

import numpy as np

temperatures_w1 = np.array([22.5,25.3,20.8,23.4,26.1,24.8,21.9])
temperatures_w1
```

```python
### 2. Inspection and Properties:
'''Inspect the shape, data type, and the number of elements in the array.
Inspect the shape of the temperatures_w1 array.'''

print("shape of array:",temperatures_w1.shape)
print("type of array:",temperatures_w1.dtype)
print("size of array:",temperatures_w1.size)
```

```python
### 3. Array Operations:
'''Convert the temperatures_w1 from Celsius to Fahrenheit using the
formula: Fahrenheit = (Celsius * 9/5) + 32.'''

fahrenheit_w1 = (temperatures_w1 * 9/5) + 32
print(fahrenheit_w1)

'''Find the maximum, minimum, and mean temperatures for the week. '''

print("maximum temperature for week:", max(temperatures_w1))
print("minimum temperature for week:", min(temperatures_w1))
print("mean temperature for week:",round(temperatures_w1.mean(),3))

```

```python
### 4. Array Slicing and Indexing:
'''Extract the temperatures for the first three days of the week. '''
print("temperatures for the first three days of week:",temperatures_w1[:3])

'''Extract the temperatures for the weekend (last two days). '''
print("temperatures for the weekend(last two days):", temperatures_w1[-2:])

'''Extract the temperatures for the middle three days of the week. '''
print("temperatures for the middle three days of the week:", temperatures_w1[2:5])
```

```python
### 5. Create a 2D Array:
temperatures = np.array([[22.5,25.3,20.8,23.4,26.1,24.8,21.9],[19.2,22.5,21.3,24.0,23.5,22.8,20.1]])
print(temperatures)
```

```python
### 6. Inspect and Slice the 2D Array:
'''Inspect and print the shape, data type, and total number of elements in the
temperatures array. '''

print("shape of array:", temperatures.shape)
print("data type of array:", temperatures.dtype)
print("number of elements:", temperatures.size)

'''Extract the temperatures for each week and the weekends (last two days)
for both weeks.'''

print("week 1 temperature:",temperatures[0])
print("week 2 temperature:",temperatures[1])
print("weekends (last two days)both weeks:",temperatures[:,-2:] )
```

```python
## Pandas Series:
```

```python
### 1. Creating Pandas Series:

import pandas as pd
marks = pd.Series([95,92,89,85,80],index=['Rank1','Rank2','Rank3','Rank4','Rank5'])
print(marks)
```

```python
### 2. Indexing and Slicing:
'''Use the integer index position to access the mark of the 1st rank student. '''

print("1st rank student mark:", marks[0])

'''Use the loc accessor to retrieve the marks of the top 3 ranks by
specifying their index labels. '''

print("\nTop 3 Ranks:",'\n',marks.loc['Rank1':'Rank3'])

'''Use the iloc accessor to retrieve the mark of the 3rd rank student. '''

print("\nMark of third rank student:", marks.iloc[2])

'''Apply a boolean mask to filter the Series and retrieve the ranks where the
marks are greater than 90.'''

marks[marks>90]
```

```python
### 3. Manipulating Series:
'''Modify the mark of the student with the 1st rank to 100.  '''
marks.loc['Rank1']=100
print(marks)
'''Remove the entry corresponding to the last rank from the Series.'''
marks.drop('Rank5',inplace=True)
print('\n',marks)
'''Compute the CGPA by dividing each mark in the Series by 10.'''
CGPA=marks/10
print('\nCGPA:','\n', CGPA)
```

```python
## Pandas DataFrame:
```

```python
### 1. Creating Pandas DataFrame:

transactions=pd.DataFrame(
    {
    'TransactionID':[101,102,103,104,105,106,107,108,109,110],
    'ProductCategory':['Electronics', 'Clothing', 'Electronics', 'Furniture',
'Clothing', 'Electronics', 'Furniture', 'Clothing', 'Furniture', 'Electronics'],
    'Region':['North', 'South', 'North', 'East', 'West', 'North', 'East', 'West',
'South', 'North'],
    'Amount':[200, 150, 300, 450, 200, 250, 300, 180, 350, 400]
}
    )
print(transactions)
```

```python
### 2. Data Exploration:
print(transactions)
print("\nInformation of dataframe:")
print(transactions.info())
print("\nHead of dataframe:")
print(transactions.head())
print("\nTail of dataframe:")
print(transactions.tail())
print("\nShape of dataframe:")
print(transactions.shape)
print("\nColumn names of dataframe:")
print(transactions.columns.tolist())
print("\nDatatype of dataframe:")
print(transactions.dtypes)
```

```python
'''Display only the 'ProductCategory' and 'Amount' columns.'''
transactions[['ProductCategory','Amount']]
```

```python
'''Retrieve the last 3 columns of the DataFrame using iloc or loc. '''
transactions.iloc[:,-3:]
```

```python
'''Filter rows where the 'Region' is 'North' and 'Amount' is greater than 200.'''
transactions[(transactions['Region']=="North")&(transactions['Amount']>200)]
```

```python
'''Find the value counts for the 'ProductCategory' column.'''
transactions['ProductCategory'].value_counts()
```

```python
'''Find the unique values in the 'Region' column. '''
transactions['Region'].unique()
```

```python
'''Group by 'Region' and find the mean amount for each region.'''
transactions.groupby('Region')['Amount'].mean()
```

```python
### 3. Manipulating the DataFrame:
```

```python
'''Modify the 'Amount' for the transaction with TransactionID 102 to 165. '''
transactions.loc[transactions['TransactionID']==102, 'Amount']=165
print(transactions)
```

```python
'''Add a new column 'Discount' by calculating 10% of the 'Amount'. '''
transactions['Discount']=transactions['Amount']*10/100
print(transactions)
```

```python
'''Remove the row with TransactionID 109. '''
transactions.drop(transactions[transactions['TransactionID']==109].index)
```

```python
'''Delete the 'Discount' column from the DataFrame'''
del(transactions['Discount'])
print(transactions)
```
