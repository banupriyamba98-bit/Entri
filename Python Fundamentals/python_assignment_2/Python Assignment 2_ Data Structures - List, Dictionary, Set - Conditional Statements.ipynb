```python
## 1. List Creation:
```

```python
### a. Create a list named age_list with five integer elements. For eg., [24, 25, 26, 27, 28]
age_list = [24,25,26,27,28]
print(age_list)
```

```python
### b. Create a list named name_list with five string elements.
name_list = ["Abi", "Banu", "Chitra", "Divya", "Ezhil"]
print(name_list)
```

```python
## 2. List Operations / Modifications:
```

```python
### a. Append the string "Yazhini" to name_list.
name_list.append("Yazhini")
print(name_list)
```

```python
### b. Insert the element 30 at index 2 in age_list.
age_list.insert(2,30)
print(age_list)
```

```python
### c. Remove the string "Yazhini" from name_list.
name_list.remove("Yazhini")
print(name_list)
```

```python
### d. Pop the last element from age_list.
age_list.pop()
print(age_list)
```

```python
### e. Extend the age_list with additional ages [29, 30, 26].
age_list.extend([29,30,26])
print(age_list)
```

```python
### f. Sort age_list in descending order.
age_list.sort(reverse=True)
print(age_list)
```

```python
### g. Find Max age, Min age and sum of all ages from age_list.
Max_age = max(age_list)
Min_age = min(age_list)
Sum_age = sum(age_list)

print(f"Maximum_age_is", Max_age)
print(f"Minimum_age_is", Min_age)
print(f"Sum_of_all_age_is", Sum_age)
```

```python
## Sets (Operations):
```

```python
### a. Create a set called my_set with following values:['a','e','i','o','u','a','a','i']Analyse the output and provide explanation for the same.

my_set = {'a','e','i','o','u','a','a','i'}
print(my_set)

## A set in python is a collection of distinct values,so it automatically removes duplicate values.
## The set will keeps only one occurence of each value
## set is unordered,it won't maintain user given order.
```

```python
### b. Attempt to change the value of my_set[4] = 's'. If code throws an error, provide an explanation.
print(my_set)
my_set[4] = 's'

## Sets are unordered,it doesn't have fixed position or index for specific element like list/tuples.
## If we try to change 4th python doesn't know the 4th element. Then it will throw the error'set' object does not support item assignment.
## We can only add and remove the elements in sets we cant modify the element.
```

```python
### c. Create two sets:  set1 with values: {1, 3, 5, 7, 9}  set2 with values: {2, 3, 5, 8, 10}
set1 = {1,3,5,7,9}
set2 = {2,3,5,8,10}
print(set1)
set2
```

```python
### d. Compute and print the union and intersection of set1 and set2.
Union_of_set = set1.union(set2)
print(f"Union = {Union_of_set}")

Intersection_of_set = set1.intersection(set2)
print(f"Intersection = {Intersection_of_set}")
```

```python
''' 2. Find the performance category based on the input score using following criteria:
a. Above Average:  Score greater than 7
b. Average: Score between 4 and 7(both inclusive)
c. Below Average: Score lesser than 4. '''

score= (int(input("Enter Your Score (0 to 10) : ")))

if score>7:
  print(f" Above Average")
elif 4 <= score <= 7:
  print(f"Average")
else:
  print(f"Below Average")
```
